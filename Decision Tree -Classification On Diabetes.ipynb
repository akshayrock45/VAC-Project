# Segregating ad reshaping the dataset

```python
y = data.iloc[:,-1].values.reshape(-1,1)
```

```python
y.shape
```

```python
y
```

```python
x = data.drop(['Outcome'],axis = 1)
```

```python
x.shape
```

```python
x
```

# Dividing the data set into training and test set

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25,random_state = 0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DC = DecisionTreeClassifier(criterion = 'entropy')
```

```python
DC.fit(x_train,y_train)
```

# Prediction

```python
y_pred = DC.predict(x_test)
```

```python
y_pred
```

# Actual vs Predicted

```python
final = pd.DataFrame(y_test,columns=['Actual'])
```

```python
final['Predicted'] = y_pred
```

```python
final
```

```python
final.groupby('Actual').size()
```

```python
final.groupby('Predicted').size()
```

```python
sns.countplot(final['Actual'],label = 'count')
```

```python
sns.countplot(final['Predicted'],label = 'count')
```
